```python
import tkinter as tk
import random

class TypingGame(tk.Frame):
    def __init__(self, master):
        super().__init__(master)
        self.master = master
        self.words = ["Once there were four children whose names were Peter, Susan, Edmund, and Lucy.","In a hole in the ground there lived a hobbit.","Here is Edward Bear, coming downstairs now, bump, bump, bump, on the back of his head, behind Christopher Robin.","All children, except one, grow up.","Marley was dead, to begin with."]
        self.current_word = ""
        self.time_remaining = 60
        self.score = 0
        self.create_widgets()

    def create_widgets(self):
        # Create labels for the game
        self.score_label = tk.Label(self.master, text="Score: {}".format(self.score))
        self.score_label.pack()

        self.timer_label = tk.Label(self.master, text="Time remaining: {}".format(self.time_remaining))
        self.timer_label.pack()

        self.word_label = tk.Label(self.master, text="Type the word:")
        self.word_label.pack()

        self.current_word_label = tk.Label(self.master, text=self.current_word)
        self.current_word_label.pack()

        # Create an entry widget for user input
        self.input_entry = tk.Entry(self.master)
        self.input_entry.pack()
        self.input_entry.bind('<Return>', self.check_word)

        # Start the game
        self.generate_word()
        self.input_entry.focus_set()
        self.countdown()

    def generate_word(self):
        self.current_word = random.choice(self.words)
        self.current_word_label.config(text=self.current_word)

    def check_word(self, event):
        user_input = self.input_entry.get()
        if user_input == self.current_word:
            self.score += 1
            self.score_label.config(text="Score: {}".format(self.score))
            self.generate_word()
            self.input_entry.delete(0, tk.END)

    def countdown(self):
        if self.time_remaining > 0:
            self.time_remaining -= 1
            self.timer_label.config(text="Time remaining: {}".format(self.time_remaining))
            self.master.after(1000, self.countdown)
        else:
            self.game_over()

    def game_over(self):
        self.input_entry.config(state='disabled')
        self.timer_label.config(text="Time's up!")
        self.word_label.config(text="Final score: {}".format(self.score))

# Create the main window
root = tk.Tk()
root.title("Typing Game")
root.geometry("400x200")

# Create and run the game
game = TypingGame(root)
game.pack()

root.mainloop()
```
